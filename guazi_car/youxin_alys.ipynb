```python
import pymongo
from bs4 import BeautifulSoup
import requests
import time
import re
import charts
```

```python
client = pymongo.MongoClient('localhost',27017)
youxin = client['youxin']
car_item = youxin['car_item']
```

```python
for car in car_item.find({},{"_id":0,'GB':1,"milestone":1,'time':1,"title":1,'price':1}).limit(20):
    print(car)
```

```python
def getpipe():
    pipeline= [
        {'$group':{'_id':'$displacement','count':{'$sum':1}}},
#         {'$sort':{'count':-1}}
    ]
    for i in car_item.aggregate(pipeline):
           yield [i['_id'],i['count']]
       # print(i)
```

```python
option={
    'chart':{'zoomType':'xy'},
    'title':{'text':'优信二手车排量比例'},
    'subtitle':{'text':'数据来源:优信二手车'},
    'plotOptions':{'pie':{'dataLabels':{'enabled':True}}}, #设置数值可见
}
series = [{
  'type'  :'pie',
    'name':'数量',
    'data':[i for i in getpipe()]
}]
charts.plot(series,options = option,show ='inline')
```

```python
for  car in car_item.find():
    title = car['title'].split()[0]
    car_item.update({'_id':car['_id']}, {'$set': {'title':  title}})
print("ok")
```

```python
def get_name():
    pipeline= [
        {'$group':{'_id':'$title',
                   'count':{'$sum':1},
                  }
        },
        {'$sort':{'count':-1}}
    ]
    for i in car_item.aggregate(pipeline):
           yield [i['_id'],
                  i['count'],
#                   i['avg_price']
                 ]
```

```python
option={
    'chart':{'zoomType':'xy'},
    'title':{'text':'优信二手品牌比例'},
    'subtitle':{'text':'数据来源:优信二手车'},
     'plotOptions':{'column':{'dataLabels':{'enabled':True}}}, #设置数值可见
}
series = [{
  'type'  :'pie',
    'name':'数量',
    'data':[i for i in get_name()]
}]
charts.plot(series,options = option,show ='inline')
```

```python
for  car in car_item.find():
    price = car['price']
    car_item.update({'_id':car['_id']}, {'$set': {'price': round(price,2)}})
print("ok")   
```
